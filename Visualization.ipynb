```python
x = np.linspace(-1,6,30) # 30 datapoints between -1 and 6
x
```

```python
y = np.linspace(-1,6,30) # 30 datapoints between -1 and 6
y
```

```python
z = x+y # For 3-D we need the z-axis
z
```

```python
ax = plt.axes(projection='3d')
ax.plot3D(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
